```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image);
```

```python
# We will work with the image above
# Convert it to grayscale
image_copy = image.copy()
image_rgb = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_rgb,cv2.COLOR_RGB2GRAY)
canny = cv2.Canny(image_gray,100,200)
# Detect the corners using the cornerHarris
dest =cv2.cornerHarris(canny,100,3,0.007)
dest = cv2.dilate(dest, None)
# Display the image with the detected corners 
canny[dest > 0.01 * dest.max()]==[0, 0, 255]
plt.imshow(canny)
```

```python
img = cv2.imread('img/chess.jpg')

# Convert img to grayscale
image_copy = img.copy()
image_rgb = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_rgb,cv2.COLOR_RGB2GRAY)
canny = cv2.Canny(image_gray,100,200)

# Detect the corners using goodFeaturesToTrack

corners = cv2.goodFeaturesToTrack(canny, 100, 0.01, 10)

# convert corners values to integer
# So that we will be able to draw circles on them
corners = np.int0(corners)


# draw red color circles on all corners
for i in corners:
    x, y = i.ravel()
    cv2.circle(image_copy, (x, y), 3, (255, 0, 0), -1)
plt.imshow(image_copy)
```
