```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('img/people2.jpg');
```

```python
image_copy = image.copy()
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

# Given the image above, lets detect faces on the image using haar cascades
# First create a CascadeClassifier using the frontalFace classifier -> Haarcascades/haarcascade_frontalface_default.xml
face_cascade = cv2.CascadeClassifier("Haarcascades/haarcascade_frontalface_default.xml")

# Convert the image to grayscale
image_gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)

# Using the classifiers detect all faces on the image
face_rect = face_cascade.detectMultiScale(image_gray,
                                              scaleFactor = 1.2,
                                              minNeighbors = 5)

# If no faces detected print("No faces found")
if len(face_rect)==0:
    print("No faces found")
else:
# Otherwise draw a rectangle on each face that has been detected
    for (x, y, w, h) in face_rect:
        cv2.rectangle(image_copy, (x, y),
                      (x + w, y + h), (0, 255, 0), 2)
plt.figure(figsize=(25,40))
plt.imshow(image_copy)
```

```python
# Repeat the process above with your own images
window_name = "filter"
cap = cv2.VideoCapture(0)
key = 0 
cv2.namedWindow(window_name,cv2.WINDOW_AUTOSIZE)

face_cascade = cv2.CascadeClassifier("Haarcascades/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("Haarcascades/haarcascade_eye.xml")
car_cascade = cv2.CascadeClassifier("Haarcascades/car.xml")

while (True):
    ret, frame = cap.read()
    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    face_rect = face_cascade.detectMultiScale(gray_frame,
                                              scaleFactor = 1.2,
                                              minNeighbors = 5)
    
    if len(face_rect)==0:
        cv2.putText(frame, 'No Face Detected', (250,50), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.5, (0,0,255), 2, cv2.LINE_AA)
    else:
    # Otherwise draw a rectangle on each face that has been detected
        for (x, y, w, h) in face_rect:
            cv2.rectangle(frame, (x, y),
                        (x + w, y + h), (0, 255, 0), 5)
            cv2.putText(frame, 'Face Detected', (x,y), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.5, (0,255,0), 2, cv2.LINE_AA)

    eye_rect = eye_cascade.detectMultiScale(gray_frame,
                                              scaleFactor = 1.2,
                                              minNeighbors = 5)


    if len(eye_rect)==0:
        cv2.putText(frame, 'No Eye Detected', (150,50), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.5, (0,0,255), 2, cv2.LINE_AA)
    else:
    
        for (x, y, w, h) in eye_rect:
            cv2.rectangle(frame, (x, y),
                        (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, 'Eyes Detected', (x,y), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.5, (0,255,0), 2, cv2.LINE_AA)


    car_rect = eye_cascade.detectMultiScale(gray_frame)



    if len(car_rect)==0:
        cv2.putText(frame, 'No Car Detected', (50,50), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.5, (0,0,255), 2, cv2.LINE_AA)
    else:

        for (x, y, w, h) in car_rect:
            cv2.rectangle(frame, (x, y),
                        (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, 'Car Detected', (x,y), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.5, (0,255,0), 2, cv2.LINE_AA)


    cv2.imshow(window_name,frame)

    if key == 27:
        break
    
    key = cv2.waitKey(1)



cap.release()

cv2.destroyAllWindows()
cv2.waitKey(1)



```
