# Bitwise operations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# With numpy create a black canvas of size 800 x 800 
black=np.zeros((800,800))

cv2.imshow("black", black)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Make a copy of the canvas and draw a white square of size 250 x 250 on top of the copy of the canvas

black_1=black.copy()

cv2.rectangle(black_1,(275,0),(525,275),(255,255,255),-1)

cv2.imshow("black", black_1)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Make a copy of the origina lcanvas and draw a white circle of size 200 x 200 on top of the copy of the canvas
# Make sure the circle is in a different position of the canvas than the square

black_2=black.copy()

cv2.circle(black_2,(400,200),200,(255,255,255),-1)

cv2.imshow("black", black_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise and operation

square_negative=cv2.bitwise_and(black_1,black_2)
plt.imshow(square_negative,cmap="gray")


```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise or operation

square_negative=cv2.bitwise_or(black_1,black_2)
plt.imshow(square_negative,cmap="gray")


```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise not operation

square_negative=cv2.bitwise_not(black_1,black_2)
plt.imshow(square_negative,cmap="gray")


```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise xor operation


square_negative=cv2.bitwise_xor(black_1,black_2)
plt.imshow(square_negative,cmap="gray")
```

```python
# Load a color image and convert it to HSV
# Identify a prominent color from the image and filter it out
# Create a mask for that color and apply a bitwise and specifying the mask to show only this part of the image
image=cv2.imread("/home/umut/Documents/GitHub/Strive_School/Computer_Vision_(Module_6)/Day_3_ Histograms_transformations_and_Bitwise_ops/img/day1.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)


lower_blue = (90, 150, 20)
upper_blue = (130, 250, 255)

mask_blue=cv2.inRange(image, lower_blue, upper_blue)

image[mask_blue==0] = [0,0,0]


plt.imshow(image)

square_negative=cv2.bitwise_not(image)
plt.imshow(square_negative,cmap="gray")

```
