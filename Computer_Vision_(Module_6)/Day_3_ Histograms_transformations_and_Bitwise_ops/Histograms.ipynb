# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc

red_flower=cv2.imread("/home/umut/Documents/GitHub/Strive_School/Computer_Vision_(Module_6)/Day_3_ Histograms_transformations_and_Bitwise_ops/img/blue-red-flowers.png")


red_f=red_flower.copy()
red_f=cv2.cvtColor(red_f,cv2.COLOR_BGR2RGB)

plt.figure(figsize = (7,5))
hist_green=cv2.calcHist([red_f],[1],None,[255],[0,250])
plt.plot(hist_green,color='g')
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
colors = ["r","g","b"]
plt.figure(figsize = (7,5))
for i,color in enumerate(colors):
    histogram=cv2.calcHist([red_f],[i],None,[255],[0,250])
    plt.plot(histogram,color=color)
plt.show()
```
