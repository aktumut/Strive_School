# Transformations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Translations

```python
# Load an image
image=cv2.imread("/home/umut/Documents/GitHub/Strive_School/Computer_Vision_(Module_6)/Day_3_ Histograms_transformations_and_Bitwise_ops/img/day2.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
image=image.copy()
```

```python
# Store the height and width of the image (you will need them later)
width,height = image.shape[:-1]
```

```python
# Create the translation Matrix
pts1 = np.float32([[50, 50],
                   [200, 50], 
                   [50, 200]])
  
pts2 = np.float32([[10, 100],
                   [200, 50], 
                   [100, 250]])
M = cv2.getAffineTransform(pts1, pts2)
```

```python
# Apply the translation to the image using cv2.warpAffine and display the new image
dst = cv2.warpAffine(image, M, (width, height))

plt.figure(figsize = (25,20))
plt.subplot(121)
plt.imshow(image)
plt.title('Input')
  
plt.subplot(122)
plt.imshow(dst)
plt.title('Output')
  
plt.show()
```

### Rotation

```python
# Load an image
image = cv2.imread("/home/umut/Documents/GitHub/Strive_School/Computer_Vision_(Module_6)/Day_3_ Histograms_transformations_and_Bitwise_ops/img/day3.jpg")
# Create a rotation matrix to rotate the image by 45 degrees (using the center as the pivot point) 
width,height = image.shape[:-1]
x_center = width/2
y_center = height/2
rotation_matrix = cv2.getRotationMatrix2D((x_center,y_center),45,1)
```

[Hint](# "cv2.getRotationMatrix2D") 


```python
# Apply the rotation matrix to the image and display the rotated image
dst = cv2.warpAffine(image, rotation_matrix , (width, height))

plt.figure(figsize = (25,20))
plt.subplot(121)
plt.imshow(image)
plt.title('Input')
  
plt.subplot(122)
plt.imshow(dst)
plt.title('Output')
  
plt.show()
```

### Scaling images

#### Interpolation methods

* INTER_NEAREST - a nearest-neighbor interpolation (Fastest)
* INTER_LINEAR - a bilinear interpolation (used by default) (good for upsampling)
* INTER_AREA - resampling using pixel area relation. It may be a preferred method for image decimation, as it gives moire’-free results. But when the image is zoomed, it is similar to the INTER_NEAREST method. (Good for downsampling)
* INTER_CUBIC - a bicubic interpolation over 4x4 pixel neighborhood (Better)
* INTER_LANCZOS4 - a Lanczos interpolation over 8x8 pixel neighborhood (Best one)

#### Downsampling
